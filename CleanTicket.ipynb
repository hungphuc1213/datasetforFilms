```python
import pandas as pd
import unicodedata
```

```python
def remove_diacritics(input_string):
    return ''.join(char for char in unicodedata.normalize('NFD', input_string) if unicodedata.category(char) != 'Mn')
def to_uppercase(input_string):
    return input_string.upper()
```

```python
for x in df[df["orderid"].isna()==1].index:
    df.drop(x, inplace=True)
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df["saledate"] = df["saledate"].astype(str)
df[["saledate", "saletime"]] = df["saledate"].str.split(expand=True)
df
```

```python
df = df.reindex(columns=['orderid', 'cashier', 'saledate', 'saletime', 'total', 'customerid', 'ticketcode', 'date'
                    , 'time', 'slot', 'room', 'film', 'slot type', 'ticket type', 'ticket price', 'popcorn'])
df
```

```python
for x in df.columns:
    df = df.rename(columns={x : x.capitalize()})
df
```

```python
gk = df.groupby("Film")
gk.count()
```

```python
df['Film'] = df['Film'].str.replace("DORAEMON: NOBITA VÀ MẶT TRĂNG PHIÊU LƯU KÝ", "Doraemon: Nobita's Chronicle of the Moon Exploration")
df['Film'] = df['Film'].str.replace("AVENGERS: HỒI KẾT", "Marvel Studios' Avengers: Endgame")
df['Film'] = df['Film'].str.replace("CHÚA TỂ GODZILLA: ĐẾ VƯƠNG BẤT TỬ", "Godzilla: King of the Monsters")
df['Film'] = df['Film'].str.replace("CORGI: NHỮNG CHÚ CHÓ HOÀNG GIA", "The Queen's Corgi")
df['Film'] = df['Film'].str.replace("JOHN WICK 3: CHUẨN BỊ CHIẾN TRANH", "John Wick: Chapter 3 – Parabellum")
df['Film'] = df['Film'].str.replace("MẸ MA THAN KHÓC LA LLORONA", "The Curse Of La Llorona")
df['Film'] = df['Film'].str.replace("NGÔI ĐỀN KỲ QUÁI", "Pee Nak")
df['Film'] = df['Film'].str.replace("NỤ HÔN MA QUÁI", "Krasue: Inhuman Kiss")
df['Film'] = df['Film'].str.replace("POKÉMON: THÁM TỬ PIKACHU", "Pokémon Detective Pikachu")
df['Film'] = df['Film'].str.replace("QUÝ CÔ LỪA ĐẢO", "The Hustle")
df['Film'] = df['Film'].str.replace("THẰNG EM LÝ TƯỞNG", "Inseparable Bros")
df['Film'] = df['Film'].str.replace("VÔ GIAN ĐẠO", "Infernal Affairs")
#Remove accents
df["Film"] = df["Film"].apply(remove_diacritics)
#Make all letters to uppercase
df["Film"] = df["Film"].apply(to_uppercase)
#gk = df.groupby("Film")
#gk.count()
```

```python
gk = df.groupby('Slot type')
gk.count()
```

```python
gk = df.groupby('Ticket type')
gk.count()
```

```python
df["Ticket type"] = df["Ticket type"].str.replace("Thành viên", "Member")
#gk = df.groupby('Ticket type')
#gk.count()
```

```python
gk = df.groupby('Popcorn')
gk.count()
```
